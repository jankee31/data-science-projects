```python
## Step 1: Simulating multiple walks
## A single random walk is one thing but that doesn't tell me if I have any chance to win the bet.
```

```python
 
import numpy as np
np.random.seed(123)
 
# Initialize all_walks 
all_walks = []
 
# Simulate random walk 10 times
for i in range(10):
 
    # Code from before
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
 
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
 
    # Append random_walk to all_walks
    all_walks.append(random_walk)
 
# Print all_walks
print(all_walks)
```

```python
## Step 2: Visualizing Walks
```

```python
import matplotlib.pyplot as plt
```

```python
 
 
# Convert all_walks to Numpy array: np_aw
np_aw=np.array(all_walks)
 
# Plot np_aw and show
plt.plot(np_aw)
plt.show()
 
# Clear the figure
plt.clf()
 
# Transpose np_aw: np_aw_t so that np_all_walks represent the position after one throw for the 10 random walks  
np_aw_t=np.transpose(np_aw)
 
# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

```python
## Step 3: Implement Clumsiness
## There's still something I forgot! I Was a bit clumsy and I have a 0.1% chance of falling down.
## Changing range 10 to 250 in top for loop.
```

```python
 
# Simulate random walk 250 times
all_walks = []
for i in range(250) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
 
        # Implement clumsiness
        if np.random.rand() <= 0.001: 
          step=0
 
        random_walk.append(step)
    all_walks.append(random_walk)
 
# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
```

```python
## Step 4: Plotting the distribution
## Knowing the end points of the random walk using histogram
```

```python
 
# Simulate random walk 500 times
all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)
 
# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
 
# Select last row from np_aw_t: ends
ends = np_aw_t[-1, :]
# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```
